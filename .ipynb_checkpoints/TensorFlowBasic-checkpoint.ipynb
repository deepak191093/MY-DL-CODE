```python
sess = tf.compat.v1.Session()  #we start the session so that now the hardware is reserved for the program (CPU, GPU, RAM, ETC).
```

```python
with tf.compat.v1.Session() as ses:

     # Build a graph.
     a = tf.constant(5.0)
     b = tf.constant(6.0)
     c = a * b
     d = tf.multiply(a,b)

     # Evaluate the tensor `c`.
     print(ses.run(c))
     print(ses.run(d))
     
```

```python
with tf.compat.v1.Session() as sess:
    hello1 = tf.constant("Hello World", name = 'hello')
    print(sess.run(hello1))
```

```python
with tf.compat.v1.Session() as sess:
    hello1 = tf.constant("Hello World", name = 'hello')
    hello2 = tf.constant("Hello World", shape = (6,2), name = 'hello2')
    print(sess.run(hello1))
    print(sess.run(hello2))
```

```python
with tf.compat.v1.Session() as S:
    hello1 = tf.constant("Hello World", shape = (), name = 'hello1')
    hello2 = tf.constant("Hello World2", shape = (2, 6), name = 'hello2')
    print(S.run([hello1, hello2]))
```

```python
with tf.compat.v1.Session() as q:
    x = tf.constant(10)
    y = tf.constant(20)
    z = tf.add(x, y)
    print(q.run(z))
```
