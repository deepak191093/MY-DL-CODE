```python
import tensorflow as tf
```

# Variables are created and tracked via the tf.Variable class. A tf.Variable represents a tensor whose value can be changed by running ops on it.

```python
# Uncomment to see where your variables get placed (see below)
# tf.debugging.set_log_device_placement(True)
```

```python
my_tensor = tf.constant([[1.0, 2.0], [3.0, 4.0]])
```

```python
my_tensor
```

```python
my_tensor.ndim
```

```python
my_tensor.shape
```

```python
my_variable = tf.Variable(my_tensor, name="deepak")
```

```python
my_variable
```

```python
my_variable.numpy()
```

# we cannot assign value once defined for tf.constant and tf.Variable

```python
bool_variable = tf.Variable([False, False, False, True])
complex_variable = tf.Variable([5 + 4j, 6 + 1j])
```

```python
bool_variable.dtype
```

```python
bool_variable.shape
```

```python
complex_variable.dtype
```

```python
complex_variable.shape
```

```python
bool_variable
```

```python
complex_variable
```

```python
tf.reduce_min(my_variable)
```

```python
tf.argmax(my_variable)
```
