```python
class Tree:
    def __init__(self, key):
        self.right = None
        self.left  = None
        self.key   = key 
```

```python
root             = Tree(1)
root.left        = Tree(2)
root.right       = Tree(3)
root.left.left   = Tree(4)
root.left.right  = Tree(5)
root.right.left  = Tree(6)
root.right.right = Tree(7)
```

```python
def preOrder(node):
    if(node!=None):
        print(node.key)
        preOrder(node.left)
        preOrder(node.right)
```

```python
preOrder(root)
```

```python
def inOrder(node):
    if(node != None):
        inOrder(node.left)
        print(node.key)
        inOrder(node.right)
```

```python
inOrder(root)
```

```python
def postOrder(node):
    if(node != None):
        postOrder(node.left)
        postOrder(node.right)
        print(node.key)
```

```python
postOrder(root)
```
