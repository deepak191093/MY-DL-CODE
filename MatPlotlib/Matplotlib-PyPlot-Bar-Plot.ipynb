```python
#importing libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
#creating data for bar plot
product_category = np.array(["Furniture", "Technology", "Office Supplies"]) #dimension
sales            = np.array([4110451.90, 4744557.50, 3787492.52]) #fact
```

```python
# add title, xlable, ylable
plt.bar(product_category, sales)
plt.title("Sales Data for a X Company")
plt.xlabel("Catogery")
plt.ylabel("Sales In $")
```

```python
#adding font dictonary for font properties
plt.bar(product_category, sales, width= 0.75)
plt.title("Sales of Company", fontdict = { "color" : "Red" , "fontsize" : 30, "fontweight" : 20, "family" : ["serif"] })
plt.xlabel("Categories",      fontdict = { "color" : "cyan", "fontsize" : 20, "fontweight" : 5,  "family" : ["arial"] })
plt.ylabel("Sales In $",      fontdict = { "color" : "pink", "fontsize" : 20, "fontweight" : 5,  "family" : ["serif"] })
```

```python
#adding color to bar and adding ticks(Scaling the value in x, y lable)
plt.bar(product_category, sales, width = 0.75, color = ["orange" , "yellow"], edgecolor = "black" )
plt.title("Sales of Company", fontdict = { "color" : "Red" , "fontsize" : 30, "fontweight" : 20, "family" : ["serif"] })
plt.xlabel("Categories",      fontdict = { "color" : "cyan", "fontsize" : 20, "fontweight" : 5,  "family" : ["arial"] })
plt.ylabel("Sales In $",      fontdict = { "color" : "pink", "fontsize" : 20, "fontweight" : 5,  "family" : ["serif"] })
yTicks = np.arange(0, 7000000, 1000000)
yLable = np.array(["0L", "10L", "20L", "30L", "40L", "50L", "60L", "70L"]) #can be generated using code.
plt.yticks(yTicks, yLable)
```
